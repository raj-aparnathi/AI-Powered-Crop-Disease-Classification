```python
import os
import random
import shutil

random.seed(42)

MASTER_PATH = os.path.join(PROJECT_PATH, "master_dataset")
FINAL_PATH = os.path.join(PROJECT_PATH, "final_dataset")

# Image extensions
IMG_EXT = (".jpg", ".jpeg", ".png")

for split in ["train", "val", "test"]:
    os.makedirs(os.path.join(FINAL_PATH, split), exist_ok=True)

for crop in os.listdir(MASTER_PATH):

    crop_path = os.path.join(MASTER_PATH, crop)

    if not os.path.isdir(crop_path):
        continue

    for disease in os.listdir(crop_path):

        disease_path = os.path.join(crop_path, disease)

        if not os.path.isdir(disease_path):
            continue

        images = [
            f for f in os.listdir(disease_path)
            if f.lower().endswith(IMG_EXT)
        ]

        random.shuffle(images)

        total = len(images)

        train_end = int(total * 0.70)
        val_end = int(total * 0.85)

        split_data = {
            "train": images[:train_end],
            "val": images[train_end:val_end],
            "test": images[val_end:]
        }

        for split_name, img_list in split_data.items():

            dst = os.path.join(
                FINAL_PATH,
                split_name,
                crop,
                disease
            )

            os.makedirs(dst, exist_ok=True)

            for img in img_list:

                shutil.copy2(
                    os.path.join(disease_path, img),
                    os.path.join(dst, img)
                )

print("✅ Dataset Split Completed Successfully!")
```

```python
import os
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

from tensorflow import keras
from tensorflow.keras import layers
```

```python
# Dataset Paths

TRAIN_DIR = r"C:\Users\91816\CropX_Project\flatten_dataset\train"
VAL_DIR   = r"C:\Users\91816\CropX_Project\flatten_dataset\val"
TEST_DIR  = r"C:\Users\91816\CropX_Project\flatten_dataset\test"

IMG_SIZE = (224, 224)
BATCH_SIZE = 32
SEED = 42

print("✅ Paths Loaded Successfully")
```

```python
train_ds = tf.keras.utils.image_dataset_from_directory(
    TRAIN_DIR,
    labels="inferred",
    label_mode="categorical",
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    shuffle=True,
    seed=SEED
)

val_ds = tf.keras.utils.image_dataset_from_directory(
    VAL_DIR,
    labels="inferred",
    label_mode="categorical",
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    shuffle=False
)

test_ds = tf.keras.utils.image_dataset_from_directory(
    TEST_DIR,
    labels="inferred",
    label_mode="categorical",
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    shuffle=False
)

class_names = train_ds.class_names
NUM_CLASSES = len(class_names)

print(f"\n✅ Total Classes: {NUM_CLASSES}")
```

```python
AUTOTUNE = tf.data.AUTOTUNE

train_ds = train_ds.prefetch(buffer_size=AUTOTUNE)
val_ds = val_ds.prefetch(buffer_size=AUTOTUNE)
test_ds = test_ds.prefetch(buffer_size=AUTOTUNE)

print("✅ Dataset Optimized")
```

```python
from tensorflow.keras.applications.mobilenet_v3 import preprocess_input

data_augmentation = keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.15),
    layers.RandomZoom(0.15),
    layers.RandomContrast(0.10),
], name="data_augmentation")

print("✅ Data Augmentation Ready")
```

```python
from tensorflow.keras.applications import MobileNetV3Large

base_model = MobileNetV3Large(
    input_shape=(224, 224, 3),
    include_top=False,
    weights="imagenet"
)

base_model.trainable = False

print("✅ Base Model Loaded")
print("Trainable:", base_model.trainable)
```

```python
inputs = keras.Input(shape=(224, 224, 3))

# Data Augmentation
x = data_augmentation(inputs)

# Official MobileNetV3 Preprocessing
x = preprocess_input(x)

# Feature Extraction
x = base_model(x, training=False)

# Classification Head
x = layers.GlobalAveragePooling2D()(x)
x = layers.BatchNormalization()(x)

x = layers.Dropout(0.4)(x)

x = layers.Dense(512, activation="relu")(x)
x = layers.BatchNormalization()(x)

x = layers.Dropout(0.3)(x)

outputs = layers.Dense(NUM_CLASSES, activation="softmax")(x)

model = keras.Model(inputs, outputs)

model.summary()
```

```python
optimizer = tf.keras.optimizers.Adam(learning_rate=1e-4)

model.compile(
    optimizer=optimizer,
    loss="categorical_crossentropy",
    metrics=[
        "accuracy",
        tf.keras.metrics.TopKCategoricalAccuracy(k=3, name="top3_accuracy")
    ]
)

print("✅ Model Compiled Successfully")
```

```python
from sklearn.utils.class_weight import compute_class_weight
import numpy as np

# Collect labels from training dataset
y_train = []

for _, labels in train_ds:
    y_train.extend(np.argmax(labels.numpy(), axis=1))

y_train = np.array(y_train)

# Compute class weights
class_weights = compute_class_weight(
    class_weight="balanced",
    classes=np.unique(y_train),
    y=y_train
)

class_weight_dict = {
    i: float(weight) for i, weight in enumerate(class_weights)
}

print("✅ Class Weights Created")
print("Total Classes:", len(class_weight_dict))

# First 10 class weights
for i in range(10):
    print(f"{i}: {class_weight_dict[i]:.3f}")
```

```python
from tensorflow.keras.callbacks import (
    ModelCheckpoint,
    EarlyStopping,
    ReduceLROnPlateau
)

import os
os.makedirs("models", exist_ok=True)

checkpoint = ModelCheckpoint(
    "models/CropX_MobileNetV3_Best.keras",
    monitor="val_accuracy",
    save_best_only=True,
    verbose=1
)

early_stop = EarlyStopping(
    monitor="val_loss",
    patience=5,
    restore_best_weights=True,
    verbose=1
)

reduce_lr = ReduceLROnPlateau(
    monitor="val_loss",
    factor=0.2,
    patience=2,
    min_lr=1e-6,
    verbose=1
)

print("✅ Callbacks Ready")
```

```python
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=5,
    class_weight=class_weight_dict,
    callbacks=[
        checkpoint,
        early_stop,
        reduce_lr
    ]
)
```

```python
# Unfreeze the base model
base_model.trainable = True

# Freeze all layers except the last 30
for layer in base_model.layers[:-30]:
    layer.trainable = False

print("Trainable layers:",
      len([layer for layer in base_model.layers if layer.trainable]))

print("Total layers:", len(base_model.layers))
```

```python
optimizer = tf.keras.optimizers.Adam(learning_rate=1e-5)

model.compile(
    optimizer=optimizer,
    loss="categorical_crossentropy",
    metrics=[
        "accuracy",
        tf.keras.metrics.TopKCategoricalAccuracy(
            k=3,
            name="top3_accuracy"
        )
    ]
)

print("✅ Fine-Tuning Model Compiled")
```

```python
history_finetune = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=15,
    initial_epoch=5,
    class_weight=class_weight_dict,
    callbacks=[
        checkpoint,
        early_stop,
        reduce_lr
    ]
)
```

```python
test_loss, test_accuracy, test_top3 = model.evaluate(test_ds)

print(f"Test Accuracy: {test_accuracy:.4f}")
print(f"Test Loss: {test_loss:.4f}")
print(f"Test Top-3 Accuracy: {test_top3:.4f}")
```
